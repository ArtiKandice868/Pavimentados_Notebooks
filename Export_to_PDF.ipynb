```python
import pandas as pd
from fpdf import FPDF
```

```python
df = pd.read_excel("CostaRica_Results.xlsx")
```

```python
df.head()
```

```python
df['Source Video File'] = df['Source Video File'].str.replace('.xlsx', '.LRV')
```

```python
df = df.drop(columns=['Intervalo Lineal Longitudinal'])
```

```python
df.head()
```

```python
fault_columns = ['Grieta Lineal Longitudinal', 
                 'Grieta Lineal Transversal', 
                 #'Intervalo Lineal Transversal', 
                 'Piel de Cocodrilo', 
                 'Protuberancia, Bache', 
                 'Desenfoque Paso Peatonal', 
                 'Desenfoque Linea Blanca', 
                 'Otras Fallas']
```

```python
df['Total Faults'] = df[fault_columns].sum(axis=1)
df.head()
```

```python
# 1. Total fault length
total_fault_length = df['Total Faults'].sum()
```

```python
# 2. Total distance analyzed
total_distance = df['section_distance'].sum()
```

```python
# 3. Percentage of faults in analyzed section
fault_percentage = (total_fault_length / total_distance) * 100
```

```python
# 4. & 5. Most and least common fault types
fault_types = df.columns[1:9]
total_faults_by_type = df[fault_types].sum()
most_common_fault = total_faults_by_type.idxmax()
most_common_fault_length = total_faults_by_type.max()
```

```python
least_common_fault = total_faults_by_type.idxmin()
least_common_fault_length = total_faults_by_type.min()
```

```python
# 6. & 7. Most and second most affected section details
sorted_sections = df.sort_values(by='Total Faults', ascending=False)

most_affected_section = sorted_sections.iloc[0]['section']
most_affected_video = sorted_sections.iloc[0]['Source Video File']
most_affected_length = sorted_sections.iloc[0]['Total Faults']

second_most_affected_section = sorted_sections.iloc[1]['section']
second_most_affected_video = sorted_sections.iloc[1]['Source Video File']
second_most_affected_length = sorted_sections.iloc[1]['Total Faults']
```

```python
# 8. Section of road in video that's most affected
video_group = df.groupby('Source Video File')['Total Faults'].sum()
most_affected_video_name = video_group.idxmax()
most_affected_video_length = video_group.max()
```

```python
# 9-14. Average fault lengths for each fault type in a 100-meter section
average_fault_lengths = df[fault_types].sum() / (total_distance / 100)
```

```python
sentences = []
sentences.append(f"In the analyzed road sections, the length for ALL faults amounted to {total_fault_length:,.2f} meters.")
sentences.append(f"The total distance analyzed is {total_distance:,.2f} meters.")
sentences.append(f"This amounts to {fault_percentage:,.2f}% of the total analyzed section.")
sentences.append(f"The most common fault type observed was '{most_common_fault}', with a total length of {most_common_fault_length:,.2f} meters across all sections analyzed.")
sentences.append(f"The least common fault type observed was '{least_common_fault}', with a total length of {least_common_fault_length:,.2f} meters across all sections analyzed.")
sentences.append(f"Section {most_affected_section} in Video {most_affected_video} was the most affected, with a cumulative fault length of {most_affected_length:,.2f} meters.")
sentences.append(f"Section {second_most_affected_section} in Video {second_most_affected_video} was the second most affected, with a cumulative fault length of {second_most_affected_length:,.2f} meters.")
sentences.append(f"The Section of Road that spans Video {most_affected_video_name} is the most affected with a cumulative fault length of {most_affected_video_length:,.2f} meters.")
for fault, avg_length in average_fault_lengths.items():
    sentences.append(f"On average, '{fault}' has a length of {avg_length:,.2f} meters in each (approximately) 100-meter section.")
```

```python
sentences_sp = []

sentences_sp.append(f"En los tramos de carretera analizados, la longitud de TODOS los defectos fue de {total_fault_length:,.2f} metros.")
sentences_sp.append(f"La distancia total analizada es de {total_distance:,.2f} metros.")
sentences_sp.append(f"Esto representa el {fault_percentage:,.2f}% del total de la sección analizada.")
sentences_sp.append(f"El tipo de fallo más común observado fue '{most_common_fault}', con una longitud total de {most_common_fault_length:,.2f} metros en todos los tramos analizados.")
sentences_sp.append(f"El tipo de fallo menos común observado fue '{least_common_fault}', con una longitud total de {least_common_fault_length:,.2f} metros en todos los tramos analizados.")
sentences_sp.append(f"La Sección {most_affected_section} en el Video {most_affected_video} fue la más afectada, con una longitud acumulada de fallos de {most_affected_length:,.2f} metros.")
sentences_sp.append(f"La Sección {second_most_affected_section} en el Video {second_most_affected_video} fue la segunda más afectada, con una longitud acumulada de fallos de {second_most_affected_length:,.2f} metros.")
sentences_sp.append(f"El tramo de carretera que abarca el Video {most_affected_video_name} es el más afectado con una longitud acumulada de fallos de {most_affected_video_length:,.2f} metros.")
for fault, avg_length in average_fault_lengths.items():
    sentences_sp.append(f"En promedio, '{fault}' tiene una longitud de {avg_length:,.2f} metros en cada tramo de (aproximadamente) 100 metros.")
```

```python
def sentences_to_pdf(sentences, output_filename):
    pdf = FPDF()
    pdf.add_page()
    
    # Setting font for the header
    pdf.set_font("Courier", 'BIU', 18)
    #pdf.cell(0, 10, "Pavimentados Summary Report", ln=True, align='C')  # center-aligned header
    pdf.cell(0, 10, "Pavimentados - Informe resumido", ln=True, align='C')  # center-aligned header
    pdf.ln(10)  # Add an additional space after the header
    
    pdf.set_font("Courier", size=14)
    
    for sentence in sentences:
        pdf.multi_cell(0, 10, sentence)
        pdf.ln(5)  # Add 5mm of space after each sentence for better readability

    pdf.output(output_filename)
```

```python
pdf_output = "CostaRica_Report_SP.pdf"
```

```python
sentences_to_pdf(formatted_sentences, pdf_output)
```
